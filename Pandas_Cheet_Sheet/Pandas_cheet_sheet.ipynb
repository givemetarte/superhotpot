```python
df2 = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                            'two'],
                    'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                    'baz': [1, 2, 3, 4, 5, 6],
                    'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
```

```python
df2
```

```python
df3 = df2.pivot(index='foo', 
          columns='bar', 
          values='baz').reset_index()
df3
```

```python
df3.melt(id_vars=['foo'], 
         value_vars=['A','B','C']).sort_values(
    ['foo','bar']).rename(columns = {'value':'baz'})
```
