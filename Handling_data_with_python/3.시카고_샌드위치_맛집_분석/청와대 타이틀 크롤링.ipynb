```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = 'https://www1.president.go.kr/petitions/best'

html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
titles = soup.find_all('a', class_='cb')
```

```python
for n in range(0,len(titles)):
    print(titles[n].get_text())
```
