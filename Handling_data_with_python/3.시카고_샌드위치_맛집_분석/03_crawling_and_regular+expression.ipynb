```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup
import openpyxl

excel_file = openpyxl.Workbook()
excel_file.remove(excel_file.active)
excel_sheet = excel_file.create_sheet('ì‚¬íšŒë‰´ìŠ¤')  # sheet ì´ë¦„ ì‘ì„± 
excel_sheet.column_dimensions['B'].width = 100   # ì»¬ëŸ¼ í¬ê¸° ì •í•˜ê¸° 

num = 0
excel_sheet.append(['ë²ˆí˜¸','ì œëª©']) #sheetì— í‘œì œëª© ë„£ê¸° 

for index in range(1,6):   # ì´ 5í˜ì´ì§€ì˜ title ê°€ì ¸ì˜¤ê¸° 
    url = 'https://www.hankookilbo.com/News/Society/HC01?Page=' + str(index)  
    page = urlopen(url)
    soup = BeautifulSoup(page, "html.parser")

    titles = soup.find_all('p', class_='title')   # ê° íƒ€ì´í‹€ì—ì„œ p ì¤‘ classê°€ titleì¸ ê²ƒ ê°€ì ¸ì˜¤ê¸° 
    for title in titles: 
        # print(title.get_text())    
        num += 1 
        excel_sheet.append([num, title.get_text()])  # íƒ€ì´í‹€ ê°œìˆ˜ì™€ íƒ€ì´í‹€ ë‚´ìš© 

cell_A1 = excel_sheet['A1']
cell_A1.alignment = openpyxl.styles.Alignment(horizontal="center")  # A1 ì–‘ì‹ centerë¡œ!
cell_B1 = excel_sheet['B1']
cell_B1.alignment = openpyxl.styles.Alignment(horizontal="center")  # B1 ì–‘ì‹ centerë¡œ!

excel_file.save('í•œêµ­ì¼ë³´ ì‚¬íšŒë‰´ìŠ¤ íƒ€ì´í‹€ í¬ë¡¤ë§.xlsx')
excel_file.close()
```
