```python
for n in pop.index:
    if pop['κ΄‘μ—­μ‹λ„'][n][-3:] not in ['κ΄‘μ—­μ‹', 'νΉλ³„μ‹', 'μμΉμ‹']:
        if pop['μ‹λ„'][n][:-1]=='κ³ μ„±' and pop['κ΄‘μ—­μ‹λ„'][n]=='κ°•μ›λ„':
            si_name[n] = 'κ³ μ„±(κ°•μ›)'
        elif pop['μ‹λ„'][n][:-1]=='κ³ μ„±' and pop['κ΄‘μ—­μ‹λ„'][n]=='κ²½μƒλ‚¨λ„':
            si_name[n] = 'κ³ μ„±(κ²½λ‚¨)'
        else:
             si_name[n] = pop['μ‹λ„'][n][:-1]
                
        for keys, values in tmp_gu_dict.items():
            if pop['μ‹λ„'][n] in values:
                if len(pop['μ‹λ„'][n])==2:
                    si_name[n] = keys + ' ' + pop['μ‹λ„'][n]
                elif pop['μ‹λ„'][n] in ['λ§μ‚°ν•©ν¬κµ¬','λ§μ‚°νμ›κµ¬']:
                    si_name[n] = keys + ' ' + pop['μ‹λ„'][n][2:-1]
                else:
                    si_name[n] = keys + ' ' + pop['μ‹λ„'][n][:-1]
        
    elif pop['κ΄‘μ—­μ‹λ„'][n] == 'μ„ΈμΆ…νΉλ³„μμΉμ‹':
        si_name[n] = 'μ„ΈμΆ…'
        
    else:
        if len(pop['μ‹λ„'][n])==2:
            si_name[n] = pop['κ΄‘μ—­μ‹λ„'][n][:2] + ' ' + pop['μ‹λ„'][n]
        else:
            si_name[n] = pop['κ΄‘μ—­μ‹λ„'][n][:2] + ' ' + pop['μ‹λ„'][n][:-1]
```
