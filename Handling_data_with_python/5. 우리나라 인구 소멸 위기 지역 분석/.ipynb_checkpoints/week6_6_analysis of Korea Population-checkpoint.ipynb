A. λ„¤ μΉ΄ν”Ό μ›λ‹μ΄ λ¨μ§€ μ•κ² λ³µμ‚¬ν•  λ‹Ήμ‹ μ •ν™•ν•κ² νλ‹¨ν•΄μ„ .copy() μµμ…μ„ μ£Όμ—μ–΄μ•Ό ν•λ”λ°... 
μ΄ κ²½μ°λ” λ°μ΄ν„°λ¥Ό μ •λ¦¬ν•λ” κ³Όμ •μ—μ„ μΌμ–΄λ‚ μΌμ΄λ―€λ΅, μ›λ‹μ€ λ‚΄κ°€ μ»¨νΈλ΅¤ κ°€λ¥ν•λ‹ λ©”μ„Έμ§€ λ³΄λ‚΄μ§€ λ§λΌλ” μλ―Έμ€μµλ‹λ‹¤. 

```python
# ν•­λ© -> κµ¬λ¶„, μ΄μΈκµ¬μ -> ν•©κ³„, λ‚¨μμΈκµ¬μ -> ν•©κ³„, μ—¬μμΈκµ¬μ -> ν•©κ³„ 
population.is_copy = False

population.rename(columns={'ν•­λ©':'κµ¬λ¶„'}, inplace=True)

population.loc[population['κµ¬λ¶„']=='μ΄μΈκµ¬μ (λ…)', 'κµ¬λ¶„'] = 'ν•©κ³„'
population.loc[population['κµ¬λ¶„']=='λ‚¨μμΈκµ¬μ (λ…)', 'κµ¬λ¶„'] = 'λ‚¨μ'
population.loc[population['κµ¬λ¶„']=='μ—¬μμΈκµ¬μ (λ…)', 'κµ¬λ¶„'] = 'μ—¬μ'
```

```python
population
```

```python
population['20-39μ„Έ'] = population['20 - 24μ„Έ'] + population['25 - 29μ„Έ'] + \
                        population['30 - 34μ„Έ'] + population['35 - 39μ„Έ']
    
population['65μ„Έμ΄μƒ'] = population['65 - 69μ„Έ'] + population['70 - 74μ„Έ'] + \
                        population['75 - 79μ„Έ'] + population['80 - 84μ„Έ'] + \
                        population['85 - 89μ„Έ'] + population['90 - 94μ„Έ'] + \
                        population['95 - 99μ„Έ'] + population['100+']
            
population.head(10)
```

```python
pop = pd.pivot_table(population, 
                     index = ['κ΄‘μ—­μ‹λ„', 'μ‹λ„'], 
                     columns = ['κµ¬λ¶„'],
                     values = ['μΈκµ¬μ', '20-39μ„Έ', '65μ„Έμ΄μƒ'])
pop
```

```python
pop['μ†λ©ΈλΉ„μ¨'] = pop['20-39μ„Έ','μ—¬μ'] / (pop['65μ„Έμ΄μƒ','ν•©κ³„'] / 2)
pop.head()
```

```python
pop['μ†λ©Έμ„κΈ°μ§€μ—­'] = pop['μ†λ©ΈλΉ„μ¨'] < 1.0
pop.head()
```
