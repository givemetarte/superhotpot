```python
import networkx as nx 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
g = nx.Graph()
```

```python
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(3,4)
g.add_edge(1,4)
g.add_edge(1,5)
g.add_edge(2,4)

# 만들 때마다 어떻게 레이블을 설정하는가에 따라 그래프 보여주는 방식이 다름. 
nx.draw(g, with_labels = True)
# nx.draw(g)
# plt.savefig('example.png')
```

```python
# a circle layout 
nx.draw_circular(g, with_labels=True)
```

```python
# 현재있는 그림을 싹 다 지우는 것
plt.clf()
```

```python
nx.draw_random(g, with_labels=True)
# 랜덤으로 그려줌 
```

```python
nx.draw_shell(g, with_labels=True)
```
