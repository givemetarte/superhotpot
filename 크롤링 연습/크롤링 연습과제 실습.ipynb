#### 크롤링 연습과제 실습 

1. 크롬 브라우저 활용해서 태그 알아내기

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://media.daum.net/economic/')
soup = BeautifulSoup(res.content,'html.parser')
items = soup.select('strong.tit_g > a.link_txt')

for item in items:
    print(item.get_text())
    
    
```
