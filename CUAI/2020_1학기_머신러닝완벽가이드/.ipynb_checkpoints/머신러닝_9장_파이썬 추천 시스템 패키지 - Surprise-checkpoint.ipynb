```python
import pandas as pd 

ratings = pd.read_csv('data/ml-latest-small/ratings.csv')
# headerì™€ indexê°€ ì‚­ì œëœ íŒŒì¼ ì €ì¥ 
ratings.to_csv('data/ml-latest-small/ratings_noh.csv', 
              index=False, header=False)
```

```python
from surprise import Reader 

# ì½ê¸° ì „ì— ë°ì´í„°ì˜ íŒŒì‹± í¬ë§·ì„ ì •ì˜í•´ì•¼ í•¨! 
# í‰ì ì˜ ë‹¨ìœ„ëŠ” 0.5, ìµœëŒ€ í‰ì ì€ 5
reader = Reader(line_format='user item rating timestamp', sep=',',
               rating_scale=(0.5, 5))
data = Dataset.load_from_file('data/ml-latest-small/ratings_noh.csv',
                             reader=reader)
```

```python
# SVD í–‰ë ¬ ë¶„í•´ ê¸°ë²•ì„ ì´ìš©í•œ ì¶”ì²œ ì˜ˆì¸¡ 
trainset, testset = train_test_split(data, test_size=.25,
                                    random_state=0)

algo = SVD(n_factors=50, random_state=0)

algo.fit(trainset)
predictions = algo.test(testset)
accuracy.rmse(predictions)
```

```python
from surprise import Reader, Dataset

ratings = pd.read_csv('data/ml-latest-small/ratings.csv')
reader = Reader(rating_scale=(0.5, 5.0))

data = Dataset.load_from_df(ratings[['userId', 'movieId','rating']], reader)
trainset, testset = train_test_split(data, 
                                    test_size=.25, random_state=0)

algo = SVD(n_factors=50, random_state=0)
algo.fit(trainset)
predictions = algo.test(testset)
accuracy.rmse(predictions)
```

- ë² ì´ìŠ¤ë¼ì¸ í‰ì  = ì „ì²´ í‰ê·  í‰ì  + ì‚¬ìš©ì í¸í–¥ ì ìˆ˜ + ì•„ì´í…œ í¸í–¥ ì ìˆ˜ 

##### êµì°¨ ê²€ì¦ê³¼ í•˜ì´í¼ íŒŒë¼ë¯¸í„° íŠœë‹ 
- êµì°¨ ê²€ì¦ê³¼ í•˜ì´í¼ íŒŒë¼ë¯¸í„° íŠœë‹ì„ ìœ„í•´ ì‚¬ì´í‚·ëŸ°ê³¼ ìœ ì‚¬í•œ cross_validate()ì™€ GridSearchCV í´ë˜ìŠ¤ë¥¼ ì œê³µí•¨. 
- cross_validate()ë¥¼ ì´ìš©í•´ ratings.csvë¥¼ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë¡œë”©í•œ ë°ì´í„°ë¥¼ 5ê°œì˜ í•™ìŠµ/ê²€ì¦ í´ë“œ ë°ì´í„° ì„¸íŠ¸ë¡œ ë¶„ë¦¬í•´ êµì°¨ ê²€ì¦ì„ ìˆ˜í–‰í•˜ê³  RMSE, MAEë¡œ ì„±ëŠ¥ í‰ê°€ë¥¼ ì§„í–‰í•¨. 

```python
from surprise.model_selection import cross_validate 

ratings = pd.read_csv('data/ml-latest-small/ratings.csv')
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 
                                    'rating']], reader)

algo = SVD(random_state=0)
cross_validate(algo, data, measures=['RMSE', 'MAE'], 
              cv=5, verbose=True)
```

```python
from surprise.model_selection import GridSearchCV

# ìµœì í™”í•  íŒŒë¼ë¯¸í„°ë¥¼ ë”•ì…”ë„ˆë¦¬ í˜•íƒœë¡œ ì§€ì • 
param_grid = {'n_epochs': [20,40,60], 
             'n_factors': [50, 100, 200]}

gs = GridSearchCV(SVD, param_grid, measures=['rmse','mae'],cv=3)
gs.fit(data)

# ìµœê³  RMSE Evaluation ì ìˆ˜ì™€ ê·¸ë•Œì˜ í•˜ì´í¼ íŒŒë¼ë¯¸í„° 
print(gs.best_score['rmse'])
print(gs.best_params['rmse'])
```

ì´ë ‡ê²Œ ì˜¤ë¥˜ê°€ ëœ¬ë‹¤ëŠ” ë§ì”€. ê·¸ë˜ì„œ í•™ìŠµ ë°ì´í„°ë¡œ ì‚¬ìš©í•˜ê¸°ìœ„í•´ DatasetAutoFolds í´ë˜ìŠ¤ë¥¼ ì´ìš©í•˜ë©´ ëœë‹¤.

```python
from surprise.dataset import DatasetAutoFolds

reader = Reader(line_format='user item rating timestamp',
               sep=',', rating_scale=(0.5, 5))
data_folds = DatasetAutoFolds(ratings_file='data/ml-latest-small/ratings_noh.csv',
                             reader=reader)

# ì „ì²´ ë°ì´í„°ë¥¼ í•™ìŠµ ë°ì´í„°ë¡œ ìƒì„±í•¨.
trainset = data_folds.build_full_trainset()
```

```python
algo = SVD(n_epochs=20, n_factors=50, random_state=0)
algo.fit(trainset)
```

```python
# ì˜í™”ì— ëŒ€í•œ ìƒì„¸ ì†ì„± ì •ë³´ ë°ì´í„°í”„ë ˆì„ ë¡œë”©
movies = pd.read_csv('data/ml-latest-small/movies.csv')

# userId=9ì˜ movieId ë°ì´í„°ë¥¼ ì¶”ì¶œí•´ movieId=42 ë°ì´í„°ê°€ ìˆëŠ”ì§€ í™•ì¸.
movieIds = ratings[ratings['userId'] == 9]['movieId']
if movieIds[movieIds == 42].count() ==0: 
    print('ì‚¬ìš©ì ì•„ì´ë”” 9ëŠ” ì˜í™” ì•„ì´ë”” 42ì˜ í‰ì  ì—†ìŒ ')
    
print(movies[movies['movieId']==42])
```

```python
#ì „ì²´ ë°ì´í„° ì˜ˆì¸¡í•˜ê¸° 
def get_unseen_surprise(ratings, movies, userId): 
    seen_movies = ratings[ratings['userId'] == userId]['movieId'].tolist()
    total_movies = movies['movieId'].tolist()
    unseen_movies = [movie for movie in total_movies if movie not in seen_movies]
    print('í‰ì  ë§¤ê¸´ ì˜í™” ìˆ˜:', len(seen_movies), 'ì¶”ì²œ ëŒ€ìƒ ì˜í™” ìˆ˜: ', 
         len(unseen_movies), 'ì „ì²´ ì˜í™” ìˆ˜:', len(total_movies))
    
    return unseen_movies

unseen_movies = get_unseen_surprise(ratings, movies, 9)
```

```python
def recomm_movie_by_surprise(algo, userId, unseen_movies, top_n=10):
    predictions = [algo.predict(str(userId), str(movieId)) 
                  for movieId in unseen_movies]
    def sortkey_est(pred):
        return pred.est
    
    predictions.sort(key=sortkey_est, reverse=True)
    top_predictions=predictions[:top_n]
    
    top_movie_ids = [ int(pred.iid) for pred in top_predictions]
    top_movie_rating = [ pred.est for pred in top_predictions]
    top_movie_titles = movies[movies.movieId.isin(top_movie_ids)]['title']
    
    top_movie_preds = [(id, title, rating) for id, title, rating 
                      in zip(top_movie_ids, top_movie_titles, 
                            top_movie_rating)]
    
    return top_movie_preds
```

```python
unseen_movies = get_unseen_surprise(ratings, movies, 9)
top_movie_preds = recomm_movie_by_surprise(algo, 9, unseen_movies, top_n=10)

print('##### Top-10 ì¶”ì²œ ì˜í™” ë¦¬ìŠ¤íŠ¸ #####')
for top_movie in top_movie_preds:
    print(top_movie[1], ':', top_movie[2])
```
