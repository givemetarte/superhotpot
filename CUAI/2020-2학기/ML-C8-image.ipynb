## Chapter8 이미지 다루기

```python
import cv2
print(cv2.__version__)
import numpy as np 
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("images/plane.jpg", cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
type(image)
```

```python
# 행렬 각 원소는 개별 픽셀에 해당
image
```

```python
# 행렬 차원 = 이미지 차원
image.shape
```

```python
# 강도는 검정색(0)부터 흰색(255)까지 
image[0,0]
```

```python
# BRG로 표현 
image_bgr = cv2.imread("images/plane.jpg", cv2.IMREAD_COLOR)
image_bgr[0,0]
```

```python
# RGB로 바꾸기 
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

```python
# 이미지 저장 
cv2.imwrite("images/plane_new.jpg", image)
```

```python
image_50x50 = cv2.resize(image, (50,50))
plt.imshow(image_50x50, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# 이미지 자르기 
image_cropped = image[:, :1800]
plt.imshow(image_cropped, cmap='gray')
plt.axis("off")
plt.show()
```

```python
# 이미지 흐리게 하기 
image_blurry = cv2.blur(image, (100,100))
plt.imshow(image_blurry, cmap='gray')
plt.axis("off")
plt.show()
```

```python
kernel = np.ones((5,5)) / 25.0
kernel
```

```python
image_kernel = cv2.filter2D(image, -1, kernel)
plt.imshow(image_kernel, cmap='gray')
plt.axis("off")
plt.show()
```
